```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# seborn

tips = sns.load_dataset('tips')

#plot
sns.scatterplot(data = tips, x="total_bill", y="tip")
plt.title("Seborn 예제")
plt.show()
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
sns.set_style("whitegrid")
sns.countplot(data = df, x='island')
plt.show()
```

```python
# 어두운 배경 스타일 설정
sns.set_style("darkgrid")
sns.countplot(data = df, x = 'island')
plt.show()
```

```python
# 그래프 요소 선 스타일 설정
sns.set_style("ticks")
sns.countplot(data = df, x = 'island')
plt.show()
```

```python
# 스타일과 폰트 스케일 설정(font_scale), 컨텍스트 설정(context)
sns.set_theme(style = "darkgrid", context = "talk",font_scale = 0.7)
# 컬러 팔레트 설정
# sns.set_palette("Set2")
sns.countplot(data = df, x='island')
plt.show()
```

```python
sns.set_palette("Set2")
sns.countplot(data = df, x='island')
plt.show()
```

```python
# 스타일과 폰트 스케일 설정(font_scale), 컨텍스트 설정(context)
sns.set_theme(style = "darkgrid", context = "talk",font_scale = 0.7)
# 컬러 팔레트 설정
sns.set_palette("Set2")
sns.countplot(data = df, x='island')
plt.show()
```

Hist Plot

```python
sns.set_style('whitegrid')
sns.histplot(data = tips, x='total_bill', y='tip')
```

```python
#KDE 히스토그램보다 더 직관적인 밀도 추정 가능
# 데이터 모양과 패턴 파악 
```

```python
sns.histplot(data = tips, x='total_bill', hue = 'sex')
```

##### Bar Plot
<font  color = "blue"> -- 범주형 데이터 평균 / 합계 표현</font>

```python
sns.barplot(x = tips['sex'], y = tips['tip'])
```

CountPlot
<font color ="blue" >범주형 데이터의 평균 </font> 

```python
sns.countplot(data = tips, x='sex')
```

Boxplot :  사분위수, 중위값, 이상치 파악

```python
sns.boxplot(data = tips, x ="total_bill")
```

```python
# x값은 smoker, y값은 total_bill
sns.boxplot(y = tips['total_bill'], x=tips['smoker'])
```

Vilolin Plot
- 데이터 분포, 변동성 시각화


```python
sns.violinplot(data = tips, x='smoker', y='total_bill')
```

```python
tips2 = tips[['total_bill', 'tip', 'size']]
tips2.corr()
```

```python
sns.clustermap(tips2.corr(), annot = True, cmap = 'viridis')
```

##### Pair Plot
- 데이터 프레임의 각 변수 쌍에 대한 산점도, 히스토그램

```python
sns.pairplot(tips2)
```

Regression Plot - 회귀선


```python
sns.regplot(data = tips, x = 'tip', y = 'total_bill')
```

LM Plot 
- 여러 조건에 따라 회귀선을 따라 고급 버전의 regression plot 

```python
sns.lmplot(data = tips, x = 'tip', y='total_bill', hue = 'sex')
```

Join Plot
- 두변수의 산점도와 각각의 히스토그램

```python
sns.jointplot(data = tips, x='total_bill', y = 'tip', kind = 'hex')
```

```python
sns.jointplot(data = tips, x='total_bill', y = 'tip', kind = 'scatter')
```
